```python
dataVendas['Data que o cliente assinou o contrato com a Hype'] = dataVendas['Data que o cliente assinou o contrato com a Hype'].str.split('-').str[1]
dataVendas['Data que o cliente assinou o contrato com a Hype'] = dataVendas['Data que o cliente assinou o contrato com a Hype'].astype(int)
dataVendas.loc[dataVendas['Data que o cliente assinou o contrato com a Hype'] == 1, 'Data que o cliente assinou o contrato com a Hype'] = 'janeiro'
dataVendas.loc[dataVendas['Data que o cliente assinou o contrato com a Hype'] == 2, 'Data que o cliente assinou o contrato com a Hype'] = 'fevereiro'
dataVendas.loc[dataVendas['Data que o cliente assinou o contrato com a Hype'] == 3, 'Data que o cliente assinou o contrato com a Hype'] = 'Março'
dataVendas.loc[dataVendas['Data que o cliente assinou o contrato com a Hype'] == 4, 'Data que o cliente assinou o contrato com a Hype'] = 'abril'
dataVendas.loc[dataVendas['Data que o cliente assinou o contrato com a Hype'] == 5, 'Data que o cliente assinou o contrato com a Hype'] = 'Maio'
dataVendas.loc[dataVendas['Data que o cliente assinou o contrato com a Hype'] == 6, 'Data que o cliente assinou o contrato com a Hype'] = 'Junho'
dataVendas.loc[dataVendas['Data que o cliente assinou o contrato com a Hype'] == 7, 'Data que o cliente assinou o contrato com a Hype'] = 'Julho'
dataVendas.loc[dataVendas['Data que o cliente assinou o contrato com a Hype'] == 8, 'Data que o cliente assinou o contrato com a Hype'] = 'agosto'
dataVendas.loc[dataVendas['Data que o cliente assinou o contrato com a Hype'] == 9, 'Data que o cliente assinou o contrato com a Hype'] = 'setembro'
dataVendas.loc[dataVendas['Data que o cliente assinou o contrato com a Hype'] == 10, 'Data que o cliente assinou o contrato com a Hype'] = 'Outubro'
dataVendas.loc[dataVendas['Data que o cliente assinou o contrato com a Hype'] == 11, 'Data que o cliente assinou o contrato com a Hype'] = 'novembro'
dataVendas.loc[dataVendas['Data que o cliente assinou o contrato com a Hype'] == 12, 'Data que o cliente assinou o contrato com a Hype'] = 'dezembro'
dataVendas
```

```python
dataVendas['classificacao_por_preco'] = list(map(lambda x: 'alto valor' if (x > 450000 and x <= 700000) else ('medio valor' if (x > 350000 and x <= 450000) else 'baixo valor'), dataVendas['Valor da proposta:']))
```

```python
dataVendas.head()
```

```python
groupPorPreco = dataVendas.groupby(dataVendas['classificacao_por_preco']).sum().sort_values(by="Valor da proposta:", ascending=False)
groupPorPreco['relevancia_por_cento'] = round((groupPorPreco['Valor da proposta:']/(groupPorPreco['Valor da proposta:'].sum()))*100,ndigits=2)
groupPorPreco[['relevancia_por_cento', 'Valor da proposta:','Valor financiado']]
```
